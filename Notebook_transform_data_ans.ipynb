```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import lit, col, split
import re
import unicodedata
from pyspark.sql.functions import col
# Initialisation de SparkSession
spark = SparkSession.builder \
    .appName("Lakehouse Data Processing with Split") \
    .getOrCreate()
 
# Chemin vers le fichier texte dans le Lakehouse
file_path = "Files/Bronze/ans/files_ans/PS_LibreAcces_Personne_activite_202501230853.txt"
colAll = split(col("value"), ";") 

raw_data = spark.read.option("header", True).option("delimiter", "|").csv(file_path)
display(raw_data)
# Fonction pour nettoyer les noms de colonnes
def clean_column_name(name):
    # Supprimer les accents
    name = ''.join(
        c for c in unicodedata.normalize('NFKD', name) if not unicodedata.combining(c)
    )
    # Appliquer le nettoyage : minuscules, remplacement des espaces et suppression des caractères spéciaux
    return (
        name.lower()  # Convertir en minuscules
        .replace(" ", "_")  # Remplacer les espaces par des underscores
        .replace("'", "_")  # Remplacer les espaces par des underscores
        .replace("(", "").replace(")", "")
        .replace(",", "").replace(";", "")
        .replace("{", "").replace("}", "")
        .replace("\n", "").replace("\t", "")
        .replace("=", "").replace("-", "_")  # Remplacer les tirets par des underscores
    )

# Appliquer la fonction à toutes les colonnes
cleaned_columns = [clean_column_name(c) for c in raw_data.columns]
raw_data = raw_data.toDF(*cleaned_columns)


print("Table Delta enregistrée avec succès ! ✅")
display(raw_data)




```

```python
# Save data in dataframe
raw_data.write.format("delta").mode("overwrite").saveAsTable("person_activitie")

print(f"Les données sont nettoyé et enregistrer en delta table")

spark.sql("SHOW TABLES").show()

spark.stop()
```
